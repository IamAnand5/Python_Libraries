### Pandas DataFrame:
#### 2D structure of data consisting of rows and columns is called dataframe. Basically it is a table like structure which has rows and columns. Here every row and column is a Series.

#### Simply, it is a table jiske har column ko series bolte hain aur row ko bhi series bolte hain.

```python
import numpy as np
import pandas as pd
```

### Creating DataFrame

```python
# using list need 2D lists

student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data, columns=['iq', 'marks','package'])
```

```python
# using dictionaries

student_dict = {
    'name' : ['A','B','C','D','X','Y'],
    'iq'  : [100,90,120,80,0,0],
    'marks' : [80,70,100,50,0,0],
    'package' : [10,7,14,20,0,0]
}

students = pd.DataFrame(student_dict)   # not need to specify the columns
students.set_index('name', inplace = True)   # set name as index and inplace do this change permanent
students
```

### Filtering a DataFrame

```python
# find all the final winners
mask = ipl['MatchNumber'] == 'Final'   # return boolean series
ipl[mask]    #display only final match
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
# single line of code
#ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many super overs finishes have occured
ipl[ipl['SuperOver'] == 'Y']

ipl[ipl['SuperOver'] == 'Y'].shape[0]

#ipl['SuperOver'].value_counts()
```

```python
# how many matches has csk won in Kolkata
# Steps:
ipl.head()
ipl['City'] == 'Kolkata'
ipl[ipl['City'] == 'Kolkata']

# Since both operations returns boolean series so bitwise AND (&) operator is used
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')]

ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage

ipl.head()
ipl['TossWinner'] == ipl['WinningTeam']
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0]*100
```

```python
# movies with rating higher than 8 and votes>10000

movies.head(2)
(movies['imdb_rating'] > 8) &  (movies['imdb_votes'] > 10000)
movies[(movies['imdb_rating'] > 8) &  (movies['imdb_votes'] > 10000)].shape[0]
```

### Important DataFrame Funcitons

```python
# astype(): given columns ka datatype change karta hai, jiss se hm memory reduce kr skte hain

ipl.info()
```

```python
ipl['ID'] = ipl['ID'].astype('int32')
```

```python
ipl.info()   #using above code we can able to reduce the memory as clearly seen in the info
```

```python
#ipl['Season'] = ipl['Season'].astype('category')
ipl['Team1'] = ipl['Team1'].astype('category')
ipl['Team2'] = ipl['Team2'].astype('category')
```

```python
ipl.info()    # datatype of Season, Team1 and Team2 changed and memory reduced to 127.4 kb from 144.9kb
```
