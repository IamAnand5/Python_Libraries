### Importing Pandas

```python
import numpy as np    # sometime with pandas it requires numpy
import pandas as pd
```

### Series from lists

```python
# string

country = ['India', 'Pakistan', "USA", 'Nepal']
pd.Series(country)    # displays index which is automatically generated and values provided by us. Also gives the dtype of values.
```

```python
# integers

runs = [23,45,21,11,0,56,119]
pd.Series(runs)    # displays autogenerated index and provided values with data types
```

```python
#custom index

marks = [56,67,89,90]
subjects = ['Physics', 'Chemistry', 'Maths','English']
print(pd.Series (marks))   # here marks are values and data type is int64 with auto indexing
```

```python
# But is we want our own index we can create that also with index attriute in pandas as shown below

print(pd.Series(marks, index = subjects))    # Now we have marks as values and subjects as index.

# if we want to change index and values then

print('\n',pd.Series(subjects, index = marks)) # now here values are subjects and index is marks
```

```python
# setting a name: we can give name to our Series object, jisko hm future me refer kr skte hain

marks = pd.Series(marks, index = subjects, name = "My Marks")   # here marks, Series class ka object hai, means Series class ke ander jo attribute hain unko marks access kr payega
print(marks)
```

### Series from dictionary

```python
# series from dict: dictionary pass krne se index automatically index ban jata hai aur values, values bn jata hai

marks = {'maths' : 90,
         'english': 80,
         'Hindi': 70,
         'Science': 75
        }
marks_series = pd.Series(marks, name = 'Mere Marks')
print(marks_series)
```

### Boolean Indexing on Series

```python
# Find number of 50's and 100's scored by Kohli

vk[vk >= 50]

vk[vk >= 50].size
```

```python
# Find the number of ducks

vk[vk == 0]

vk[vk == 0].size
```

```python
# Count the number of day when I had more than 200 subs a day

subs[subs > 200]

subs[subs > 200].size
```

```python
# Find actors who have done more than 20 movies

num_movies = movies.value_counts()
num_movies[num_movies > 20]
```
